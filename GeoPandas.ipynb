[GeoPandas](https://geopandas.org/en/stable/index.html) : The goal of GeoPandas is to make working with geospatial data in python easier. It combines the capabilities of pandas and shapely, providing geospatial operations in pandas and a high-level interface to multiple geometries to shapely. GeoPandas enables you to easily do operations in python that would otherwise require a spatial database such as PostGIS.

````python
# Returns path of a particular map
geopandas.datasets.get_path('naturalearth_lowres')
````

````python
# Opens the map as a GeoDataFrame
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

#Exploring the dataset

print(world.info())

world
````

````python
# Subsets the GeoDataFrame
usa = world[world.name == "United States of America"]

# Plots
world.plot()

usa.plot()
````

````python
#creating Africa subset
africa = world[world.continent == 'Africa']

#creating a new population column with population in millions
africa['pop_millions'] = africa['pop_est'] / 1000000

#checking to make sure that everything worked out :) 
africa.head()
````

````python
#creating the plot
africa.plot(column = 'pop_millions', #column corresponds to the color
            legend = True, 
            legend_kwds = {'label': "Population by Country (in millions)",
                          'orientation': "horizontal"}) #customizing the legend

plt.title("Population of Countries in Africa") #adding a title
plt.ylabel("Longitude") #adding y axis label
plt.xlabel("Latitude") #adding x axis label
````
